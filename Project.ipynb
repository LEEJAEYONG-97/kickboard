```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import os
from datetime import datetime
import pytz  # 타임존 처리를 위한 라이브러리

# 모델 로드
model = YOLO('/content/yolov9/runs/detect/train4/weights/best.pt')  # 학습된 모델 경로로 변경

# 비디오 파일 경로 및 출력 경로 설정
video_path = r'/content/sample.mp4'
output_video_path = r'/content/output_video.avi'
capture_dir = r'/content/captured_images/'

# 디렉토리 존재 확인 및 생성
os.makedirs(capture_dir, exist_ok=True)

# 비디오 파일 열기
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Unable to open video file at {video_path}")
    exit()

# 비디오 작성기 설정
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

# 캡처 플래그 및 카운터 초기화
capture_done = False
capture_counter = 0

# 한국 타임존 설정
kst = pytz.timezone('Asia/Seoul')

# 프레임 카운터 초기화
frame_count = 0
fps = cap.get(cv2.CAP_PROP_FPS)  # 비디오의 FPS 가져오기

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 현재 시간 가져오기 (한국 시간)
    current_time = datetime.now(kst).strftime('%Y-%m-%d %H:%M:%S')

    # 현재 재생 시간 계산 (초 단위)
    play_time_seconds = int(frame_count / fps)
    play_time_str = f'{play_time_seconds // 3600:02}:{(play_time_seconds % 3600) // 60:02}:{play_time_seconds % 60:02}'

    # 프레임을 모델에 입력
    results = model(frame)

    # 바운딩 박스 개수 초기화
    bounding_box_count = 0

    # 결과 처리 및 비디오 프레임에 오버레이
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # 바운딩 박스 좌표
        confidences = result.boxes.conf.cpu().numpy()  # 신뢰도
        classes = result.boxes.cls.cpu().numpy()  # 클래스

        for box, conf, cls in zip(boxes, confidences, classes):
            if conf >= 0.8:
                x1, y1, x2, y2 = map(int, box)
                label = f'{model.names[int(cls)]} {conf:.2f}'

                # 객체를 프레임에 그리기
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)  # 폰트 크기 및 두께 증가

                # 바운딩 박스 카운트 증가
                bounding_box_count += 1

                # 조건을 만족하고 캡처가 아직 진행되지 않았을 때
                if not capture_done and conf >= 0.8:
                    capture_counter += 1
                    obj_image_path = os.path.join(capture_dir, f'captured_object_{capture_counter}.png')

                    # 캡처할 프레임을 복사하여 저장할 이미지 생성
                    captured_frame = frame.copy()

                    # 캡처본에 시간과 바운딩 박스 개수 표시
                    obj_current_time = current_time
                    text = f'Time: {obj_current_time} | Boxes: {bounding_box_count}'
                    cv2.putText(captured_frame, text, (10, captured_frame.shape[0] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # 폰트 크기 및 두께 증가

                    cv2.imwrite(obj_image_path, captured_frame)

                    # 캡처 완료 플래그 설정
                    capture_done = True
                    print(f"Captured image saved to {obj_image_path}")

    # 현재 시간과 바운딩 박스 개수 표시 (상단)
    text = f'Time: {current_time} | Boxes: {bounding_box_count}'
    cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # 폰트 크기 및 두께 증가, 위치를 상단으로 조정

    # 플레이 타임 표시 (하단)
    play_time_text = f'Play Time: {play_time_str}'
    text_size, _ = cv2.getTextSize(play_time_text, cv2.FONT_HERSHEY_SIMPLEX, 1.0, 2)
    text_w, text_h = text_size
    cv2.rectangle(frame, (10, frame.shape[0] - text_h - 20), (10 + text_w, frame.shape[0] - 10), (0, 0, 0), thickness=cv2.FILLED)
    cv2.putText(frame, play_time_text, (10, frame.shape[0] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # 폰트 크기 및 두께 증가, 위치를 하단으로 조정

    # 결과를 비디오 파일로 저장
    out.write(frame)

    # 화면에 표시 (디버깅용)
    cv2_imshow(frame)

    # 'q'를 눌러 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_count += 1  # 프레임 카운터 증가

# 자원 해제
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video saved to {output_video_path}")

```
