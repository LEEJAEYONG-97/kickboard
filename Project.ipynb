```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import os
from datetime import datetime
import pytz  # íƒ€ì„ì¡´ ì²˜ë¦¬ë¥¼ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬

# ëª¨ë¸ ë¡œë“œ
model = YOLO('/content/yolov9/runs/detect/train4/weights/best.pt')  # í•™ìŠµëœ ëª¨ë¸ ê²½ë¡œë¡œ ë³€ê²½

# ë¹„ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ë° ì¶œë ¥ ê²½ë¡œ ì„¤ì •
video_path = r'/content/sample.mp4'
output_video_path = r'/content/output_video.avi'
capture_dir = r'/content/captured_images/'

# ë””ë ‰í† ë¦¬ ì¡´ì¬ í™•ì¸ ë° ìƒì„±
os.makedirs(capture_dir, exist_ok=True)

# ë¹„ë””ì˜¤ íŒŒì¼ ì—´ê¸°
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Unable to open video file at {video_path}")
    exit()

# ë¹„ë””ì˜¤ ì‘ì„±ê¸° ì„¤ì •
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

# ìº¡ì²˜ í”Œë˜ê·¸ ë° ì¹´ìš´í„° ì´ˆê¸°í™”
capture_done = False
capture_counter = 0

# í•œêµ­ íƒ€ì„ì¡´ ì„¤ì •
kst = pytz.timezone('Asia/Seoul')

# í”„ë ˆì„ ì¹´ìš´í„° ì´ˆê¸°í™”
frame_count = 0
fps = cap.get(cv2.CAP_PROP_FPS)  # ë¹„ë””ì˜¤ì˜ FPS ê°€ì ¸ì˜¤ê¸°

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # í˜„ì¬ ì‹œê°„ ê°€ì ¸ì˜¤ê¸° (í•œêµ­ ì‹œê°„)
    current_time = datetime.now(kst).strftime('%Y-%m-%d %H:%M:%S')

    # í˜„ì¬ ì¬ìƒ ì‹œê°„ ê³„ì‚° (ì´ˆ ë‹¨ìœ„)
    play_time_seconds = int(frame_count / fps)
    play_time_str = f'{play_time_seconds // 3600:02}:{(play_time_seconds % 3600) // 60:02}:{play_time_seconds % 60:02}'

    # í”„ë ˆì„ì„ ëª¨ë¸ì— ì…ë ¥
    results = model(frame)

    # ë°”ìš´ë”© ë°•ìŠ¤ ê°œìˆ˜ ì´ˆê¸°í™”
    bounding_box_count = 0

    # ê²°ê³¼ ì²˜ë¦¬ ë° ë¹„ë””ì˜¤ í”„ë ˆì„ì— ì˜¤ë²„ë ˆì´
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œ
        confidences = result.boxes.conf.cpu().numpy()  # ì‹ ë¢°ë„
        classes = result.boxes.cls.cpu().numpy()  # í´ë˜ìŠ¤

        for box, conf, cls in zip(boxes, confidences, classes):
            if conf >= 0.8:
                x1, y1, x2, y2 = map(int, box)
                label = f'{model.names[int(cls)]} {conf:.2f}'

                # ê°ì²´ë¥¼ í”„ë ˆì„ì— ê·¸ë¦¬ê¸°
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)  # í°íŠ¸ í¬ê¸° ë° ë‘ê»˜ ì¦ê°€

                # ë°”ìš´ë”© ë°•ìŠ¤ ì¹´ìš´íŠ¸ ì¦ê°€
                bounding_box_count += 1

                # ì¡°ê±´ì„ ë§Œì¡±í•˜ê³  ìº¡ì²˜ê°€ ì•„ì§ ì§„í–‰ë˜ì§€ ì•Šì•˜ì„ ë•Œ
                if not capture_done and conf >= 0.8:
                    capture_counter += 1
                    obj_image_path = os.path.join(capture_dir, f'captured_object_{capture_counter}.png')

                    # ìº¡ì²˜í•  í”„ë ˆì„ì„ ë³µì‚¬í•˜ì—¬ ì €ì¥í•  ì´ë¯¸ì§€ ìƒì„±
                    captured_frame = frame.copy()

                    # ìº¡ì²˜ë³¸ì— ì‹œê°„ê³¼ ë°”ìš´ë”© ë°•ìŠ¤ ê°œìˆ˜ í‘œì‹œ
                    obj_current_time = current_time
                    text = f'Time: {obj_current_time} | Boxes: {bounding_box_count}'
                    cv2.putText(captured_frame, text, (10, captured_frame.shape[0] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # í°íŠ¸ í¬ê¸° ë° ë‘ê»˜ ì¦ê°€

                    cv2.imwrite(obj_image_path, captured_frame)

                    # ìº¡ì²˜ ì™„ë£Œ í”Œë˜ê·¸ ì„¤ì •
                    capture_done = True
                    print(f"Captured image saved to {obj_image_path}")

    # í˜„ì¬ ì‹œê°„ê³¼ ë°”ìš´ë”© ë°•ìŠ¤ ê°œìˆ˜ í‘œì‹œ (ìƒë‹¨)
    text = f'Time: {current_time} | Boxes: {bounding_box_count}'
    cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # í°íŠ¸ í¬ê¸° ë° ë‘ê»˜ ì¦ê°€, ìœ„ì¹˜ë¥¼ ìƒë‹¨ìœ¼ë¡œ ì¡°ì •

    # í”Œë ˆì´ íƒ€ì„ í‘œì‹œ (í•˜ë‹¨)
    play_time_text = f'Play Time: {play_time_str}'
    text_size, _ = cv2.getTextSize(play_time_text, cv2.FONT_HERSHEY_SIMPLEX, 1.0, 2)
    text_w, text_h = text_size
    cv2.rectangle(frame, (10, frame.shape[0] - text_h - 20), (10 + text_w, frame.shape[0] - 10), (0, 0, 0), thickness=cv2.FILLED)
    cv2.putText(frame, play_time_text, (10, frame.shape[0] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2)  # í°íŠ¸ í¬ê¸° ë° ë‘ê»˜ ì¦ê°€, ìœ„ì¹˜ë¥¼ í•˜ë‹¨ìœ¼ë¡œ ì¡°ì •

    # ê²°ê³¼ë¥¼ ë¹„ë””ì˜¤ íŒŒì¼ë¡œ ì €ì¥
    out.write(frame)

    # í™”ë©´ì— í‘œì‹œ (ë””ë²„ê¹…ìš©)
    cv2_imshow(frame)

    # 'q'ë¥¼ ëˆŒëŸ¬ ì¢…ë£Œ
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_count += 1  # í”„ë ˆì„ ì¹´ìš´í„° ì¦ê°€

# ìì› í•´ì œ
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video saved to {output_video_path}")

```
